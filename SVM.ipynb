# simple SVM
svm to predict if a student pass or fail based on the student's pointer or gpa

if gpa > 5 then pass else fail

ex:

gpa -> 5 then fail

gpa -> 6 then pass

gpa -> 2 then fail

and so on

```python
from sklearn import svm
# if X > 5 then P else F
X = [ [5], [6], [2], [7], [9]] # features (pointers/gpa)
y = ['F', 'P', 'F', 'P', 'P'] # lables (result)
clf = svm.SVC(kernel='linear').fit(X,y) # fit it in SVM
clf.predict([ [4] ]) # output -> F

```

```python
clf.predict([ [8] ]) # output -> P
```

#SVM code of iris in one place

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = sns.load_dataset('iris')

X = data.iloc[:, :-1] # independent var i.e. X = data['sepal_length' 'sepal_width', 'petal_length', 'petal-width' ]
y = data.iloc[:, -1] # dependent var i.e. y = data['species']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=99)

from sklearn.svm import SVC
model = SVC(kernel='linear', C=1.0, random_state=5)
model.fit(X_train, y_train)

pred = model.predict(X_test)
print(pred)
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, pred))
```
