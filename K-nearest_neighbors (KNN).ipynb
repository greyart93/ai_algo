```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.5, random_state=55)
knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

mismatched = (y_pred != y_test).sum()
matched = (y_pred == y_test).sum()
print("Mismatched:", mismatched)
print("Matched:", matched)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
scores = []
for i in range(1, 9):
  X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.5, random_state=55)
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train, y_train)
  y_pred = knn.predict(X_test)
  accuracy = accuracy_score(y_test, y_pred)
  scores.append(accuracy)
```

```python
scores # 18, 55, 70, 73,79, 87, 91 -> 0.986
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,10),scores)
```
