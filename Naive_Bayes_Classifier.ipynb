```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.5, random_state=42)

gnb = GaussianNB()
gnb.fit(X_train, y_train)

y_pred = gnb.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

# solution (only formula):

P(Yes|Sunny,High,False) = P(Sunny | Yes) *  P(High|Yes) * P(False|Yes) * P(Yes)

P(No | Sunny,High,False) = P(Sunny | No) *  P(High|No) * P(False|No) * P(No)

then compare and decide using the maximum a posteriori rule

basically P(Yes) > P(No) then P=Yes else P=No
