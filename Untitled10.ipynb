```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import requests
from io import StringIO
```

```python
# Load dataset
url = "https://gist.githubusercontent.com/greyart93/bb3dd1d009a87412d7219b0952669e1a/raw/05983ebaad0b3de1fbb0f431047f37a5a76097ac/play_tennis.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

df = pd.read_csv(data)
```

```python
df
```

```python
# Preprocess data
df['outlook'] = df['outlook'].map({'overcast': 0, 'sunny': 1, 'rainy': 2})
df['humidity'] = df['humidity'].map({'high': 0, 'normal': 1})
df['windy'] = df['windy'].map({False: 0, True: 1})
df['play'] = df['play'].map({'no': 0, 'yes': 1})

X = df[['outlook', 'humidity', 'windy']]
y = df['play']
```

```python
df
```

```python
X
```

```python
y
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=5)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

data is splitted from the root node i.e. outlook
outlook < 0.5 (overcast)
if -> true -> play=yes or class=yes i.e. 1

else -> false -> check humidity
humidity < 0.5 (0 -> high) -> split the data on this condition i.e. outlook >1.5 i.e. sunny so if sunny has humidity < 0.5 and if
true -> play=0
false -> play=1

then check windy: if windy <0.5 (false) then
true -> play=yes=1
false -> play=no=0



```python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10, 10

plot_tree(clf, feature_names=X.columns, class_names=['no', 'yes'], filled=True)
```
